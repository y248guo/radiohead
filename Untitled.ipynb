## Scraping from azlyrics

```python
import requests
```

```python
url_base = "https://www.azlyrics.com/lyrics/radiohead/"
song_title = "howdoyou"
URL = url_base + song_title + ".html"
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find(id="cf_text_top")
```

```python
print(results.prettify())
```
